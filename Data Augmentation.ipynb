Since we have only 40 images for each of the classes which is very less to train a deep learning model.<br />
We can increase the number of images by Data augmentation technique.<br />
Data Augmentation is a technique by which mulltiple images are generated by flipping,rotating, changing brightness level etc of the original image.


```python
#lets import necesasry libraries
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
```

```python
from keras.preprocessing.image import ImageDataGenerator,img_to_array,array_to_img,load_img
```

```python
datagen=ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
import os
from os import listdir
```

```python
#lets make approximately 1000 augmented copies of each image in Bacterial leaf blight

filepath='D:/Datamites/Projects/Rice Leaf/Bacterial leaf blight'


for images in os.listdir(filepath): #iterating over the images in the folder
    image=filepath+'/'+images #specifying the filepath for each image
    img=load_img(image)
    x=img_to_array(img)
    x=x.reshape((1,) + x.shape)
    i=0
    for batch in datagen.flow(x,batch_size=1,
                              save_to_dir='Bacterial leaf blight',save_prefix='DSC',save_format='jpg'):   #saving the image to original folder as 'jpg' format
        i+=1
        if i>37:   #we need 37 augmented images for each of the image
            break
```

```python
#Now same for the Brown spot

filepath='D:/Datamites/Projects/Rice Leaf/Brown spot'


for images in os.listdir(filepath): #iterating over the images in the folder
    image=filepath+'/'+images #specifying the filepath for each image
    img=load_img(image)
    x=img_to_array(img)
    x=x.reshape((1,) + x.shape)
    i=0
    for batch in datagen.flow(x,batch_size=1,
                              save_to_dir='Brown spot',save_prefix='ESC',save_format='jpg'):   #saving the image to original folder as 'jpg' format
        i+=1
        if i>37:   #we need 37 augmented images for each of the image
            break
```

```python
#Lets repeat the  same for the Leaf smut

filepath='D:/Datamites/Projects/Rice Leaf/Leaf smut'


for images in os.listdir(filepath): #iterating over the images in the folder
    image=filepath+'/'+images #specifying the filepath for each image
    img=load_img(image)
    x=img_to_array(img)
    x=x.reshape((1,) + x.shape)
    i=0
    for batch in datagen.flow(x,batch_size=1,
                              save_to_dir='Leaf smut',save_prefix='FSC',save_format='jpg'):   #saving the image to original folder as 'jpg' format
        i+=1
        if i>37:   #we need 37 augmented images for each of the image
            break
```
